```python
#Filtro de filas vs columnas
df = pd.read_csv(io.BytesIO(uploaded['file.csv']), index_col=2)

# Para aplicar este filtro SI se requiere tener una columna como índice
#df = pd.read_csv(io.BytesIO(uploaded['file.csv']), index_col=2)

filtro5 = df.loc[["Puebla","Durango"], ["Poblacion_ocupada_en_la_industria_manufacturera"]]
filtro5
```

```python
#Filtro por cabecera
filtro6 = df.head()
filtro6
```

```python
#FILTROS BASICOS PARA VARIABLES CUANTITATIVAS
#Filtro por comparación
filtro7=df[df["Periodo"] >= 2006]
filtro7
```

```python
#FILTROS BASICOS CON CONECTORES LOGICOS
#Filtro Y
filtro8 = df[(df["Periodo"] > 2006) & (df["Sexo"] == "Mujeres")]
filtro8
```

```python
#Filtro O
#Se puede cumplir una u otra clausula
filtro9 = df[ (df["Periodo"] > 2020) | (df["Sexo"] == "Hombre")]
filtro9
```

```python
#Filtro NOT
filtro10 = df[~(df["Periodo"] >= 2006) & (df["Sexo"] == "Mujeres")]
filtro10
```

```python
#Convertir archivo filtrado a CSV
filtro9.to_csv("Entregable1.csv")
```
